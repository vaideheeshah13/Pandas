```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "/Users/Vaidehee/nu-chi-data-pt-02-2021-u-c-master 3/02-Homework/04-Pandas/Instructions/HeroesofPymoli/Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

```python
#total players using LEN
total_players = len(purchase_data["SN"].value_counts())
total_players

#create data frame w total players name

total_player_count = pd.DataFrame({"Total Player":[total_players]})
total_player_count
```

```python
## Purchasing Analysis (Total)
```

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#calculations
#number of unique items

number_of_unique_items = len((purchase_data["Item ID"]).unique())

#average price

average_price = (purchase_data["Price"]).mean()

#number of purchases

number_of_purchases = (purchase_data["Purchase ID"]).count()

#total revenue

total_revenue = (purchase_data["Price"]).sum()

#create dataframe

purchasing_analysis = pd.DataFrame({"Number of Unique Items": [number_of_unique_items],
                  "Average_Price": [average_price],
                  "Number of Purchases": [number_of_purchases],
                  "Total Revenue": [total_revenue]})

#display Data Frame
purchasing_analysis
```

```python
## Age Demographics
```

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999 ]
age_categories =["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39","40+" ]

#categorize the existing players w the age bins

purchase_data["Age Group"] = pd.cut(purchase_data["Age"],age_bins, labels = age_categories)

#calculations
#total age

total_age = purchase_data["Age Group"].value_counts()

#percentage by age
percentage_by_age = (total_age / total_players) * 100

#create data frame
age_demographics = pd.DataFrame({"Percentage of Players": percentage_by_age,
                                 "Total Count": total_age})

#round the percentage column to 2 decimal points

age_demographics = age_demographics.round(2)

#sort index column
age_demographics.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#bin the purchase_data data frame by age
purchase_data["Age Group"] = pd.cut(purchase_data["Age"],age_bins, labels = age_categories)

#calculations
#purchase_count

purchase_count = purchase_data.groupby(["Age Group"]).count()["Price"]

#average purchase price

average_purchase_price = purchase_data.groupby(["Age Group"]).mean()["Price"]

#total purchase value

total_purchase_value = purchase_data.groupby(["Age Group"]).sum()["Price"]

#average purchase total 

average_purchase_total = total_purchase_value / total_age

#create dataframe

age_demographics = pd.DataFrame({"Purchase Count": purchase_count,
                                "Average Purchase Count": average_purchase_price,
                                "Total Purchase Value": total_purchase_value,
                                "Average Purchase Total Per Person": average_purchase_total})
#display data frame
age_demographics

```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Identify the the top 5 spenders in the game by total purchase value, then list (in a table):
# SN
#purchase Count
#Average Purchase Price
#Total Purchase Value

#group purchase data by SN

spender_stats = purchase_data.groupby(["SN"])

#calculations
#purchase count by spender 

spender_purchase_count = spender_stats["Purchase ID"].count()

#average purchase price per spender

avg_purchase_price_spender = spender_stats["Price"].mean()

#total purchase

total_spender_purchase_price = spender_stats["Price"].sum()

# create data frame

top_spender = pd.DataFrame({"Purchase Count": spender_purchase_count,
                                "Average Purchase Price": avg_purchase_price_spender,
                                "Total Purchase Value": total_spender_purchase_price})

#sort total purchase values in descending order                                
sorted_top_sender = top_spender.sort_values("Total Purchase Value", ascending = False)

#display data frame                                
sorted_top_sender
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Identify the 5 most popular items by purchase count, then list (in a table):
#Item ID
#Item Name
#Purchase Count
#Item Price
#Total Purchase Value

#retrieve item id, item name and item price columns

item_information = purchase_data[["Item ID", "Item Name", "Price"]]

#group item information by item name and item id

item_data = item_information.groupby(["Item ID", "Item Name"])

#calculations

#purchase count

purchase_count = item_data["Price"].count()

#average price per item

average_item_price = item_data["Price"].mean()

#total purchase value

total_purchase_value = item_data["Price"].sum()

#create data frame

most_popular_items = pd.DataFrame({"Purchase Count": purchase_count,
                                   "Item Price": average_item_price,
                                   "Total Purchase Value": total_purchase_value})

#sort purchase count by descending order

sorted_most_popular_items = most_popular_items.sort_values(["Purchase Count"], ascending=False)

#display table 

sorted_most_popular_items
```
